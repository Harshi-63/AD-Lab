```python
import google.generativeai as genai
from sentence_transformers import SentenceTransformer
import faiss

# Set up Gemini API Key
genai.configure(api_key="AIzaSyDpqHUCWQZKasQ7OEWTWjxVVxPYOAnV2jg")

# Load embedding model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF file."""
    reader = PdfReader(pdf_path)
    text = "\n".join([page.extract_text() for page in reader.pages if page.extract_text()])
    return text

def split_text(text, chunk_size=500, overlap=50):
    """Splits text into chunks for processing."""
    splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=overlap)
    return splitter.split_text(text)

def get_gemini_response(prompt):
    """Generates a response using Gemini API."""
    model = genai.GenerativeModel("gemini-1.5-pro")
    response = model.generate_content(prompt)
    return response.text if hasattr(response, "text") else "No response generated."

def embed_chunks(chunks):
    """Creates embeddings for text chunks using SentenceTransformer and FAISS."""
    embeddings = embedding_model.encode(chunks)
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings)
    return index, embeddings

def search_similar_chunks(index, embeddings, query, chunks, k=3):
    """Finds the top-k similar chunks based on cosine similarity."""
    query_embedding = embedding_model.encode([query])
    _, top_k_indices = index.search(query_embedding, k)
    return [chunks[i] for i in top_k_indices[0]]

def analyze_resume(chunks, analyze):
    """Performs resume analysis using Gemini API and FAISS search."""
    index, embeddings = embed_chunks(chunks)
    docs = search_similar_chunks(index, embeddings, analyze, chunks)
    combined_text = "\n\n".join(docs)
    prompt = f"{analyze}\n\nResume Content:\n{combined_text}"
    return get_gemini_response(prompt)

def resume_summary(query_with_chunks):
    return f"Provide a detailed summary of the resume:\n\n{query_with_chunks}"

def resume_strength(query_with_chunks):
    return f"Analyze and explain the strengths of this resume:\n\n{query_with_chunks}"

def resume_weakness(query_with_chunks):
    return f"Analyze the weaknesses of this resume and suggest improvements:\n\n{query_with_chunks}"

def job_title_suggestion(query_with_chunks):
    return f"Based on this resume, suggest job roles that can be applied on LinkedIn:\n\n{query_with_chunks}"

# Load resume from PDF
resume_text = extract_text_from_pdf(r"C:\Users\KIIT\Documents\GitHub\AD-Lab\Open_Lab-1\Angie 1Col Neat CS BTech Sample CV 2025.pdf")  # Provide the correct path
chunks = split_text(resume_text)

# Process Resume
summary = resume_summary("\n".join(chunks))
summary_result = analyze_resume(chunks, summary)
print("\n--- Resume Summary ---\n", summary_result)

strength = resume_strength(summary_result)
strength_result = analyze_resume(chunks, strength)
print("\n--- Resume Strengths ---\n", strength_result)

weakness = resume_weakness(summary_result)
result_weakness = analyze_resume(chunks, weakness)
print("\n--- Resume Weaknesses ---\n", result_weakness)

suggestion = job_title_suggestion(summary_result)
result_suggestion = analyze_resume(chunks, suggestion)
print("\n--- Job Title Suggestions ---\n", result_suggestion)
```
